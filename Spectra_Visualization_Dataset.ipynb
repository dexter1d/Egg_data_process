```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the spectral data
file_path = '/content/spectra and reference parameters.csv'  # Change the file path here
spectra_df = pd.read_csv(file_path)

# Extract wavelengths and spectra
wavelengths = spectra_df.columns.astype(float)
spectra = spectra_df.values

# Deep, high-saturation color palette
deep_colors = [
    '#8B0000', '#0000CD', '#006400', '#FF8C00', '#800080',
    '#008B8B', '#DC143C', '#228B22', '#00008B', '#B22222',
    '#8B008B', '#2F4F4F', '#9932CC'
]

# Plot
plt.figure(figsize=(6, 4), dpi=800)  ############ Change plot size and  dpi here########
num_colors = len(deep_colors)

for i in range(spectra.shape[0]):
    color = deep_colors[i % num_colors]
    plt.plot(wavelengths, spectra[i], color=color, linewidth=1)

# Axis styling
plt.xlabel('Wavelength (nm)', fontsize=14, weight='bold')
plt.ylabel('DN value', fontsize=14, weight='bold')
plt.xticks(fontsize=12, weight='bold')
plt.yticks(fontsize=12, weight='bold')
plt.xlim([wavelengths.min(), wavelengths.max()])
plt.ylim([0, spectra.max() + 100])
plt.grid(False)

# Thicker border lines
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.5)

plt.tight_layout()
plt.show()
```
